```python
from ultralytics import YOLO
import os
import csv
import numpy as np

# Paths
model_path = r"C:\Users\USER\Desktop\ant\runs\detect\new\additional_training\weights\epoch100.pt"
penguin_list_path = r"C:\Users\USER\Desktop\ant\ant_image_names.txt"
image_dir = r"C:\Users\USER\Downloads\FSC147_384_V2\images_384_VarV2"
output_csv = r"C:\Users\USER\Desktop\penguin\yolopredict.csv"
ground_truth_csv = r"C:\Users\USER\Desktop\ant\ant_gt_counts.csv"

# Load the YOLO model
model = YOLO(model_path)

# Load the list of penguin image names
with open(penguin_list_path, 'r') as file:
    penguin_image_names = [line.strip() for line in file]

# Predict penguin counts and save results
predictions = []

for image_name in penguin_image_names:
    image_path = os.path.join(image_dir, image_name)
    
    if os.path.exists(image_path):
        # Perform prediction
        results = model.predict(source=image_path)
        
        # Count detected penguins (assumes class "penguin" is labeled correctly in the model)
        predicted_count = len(results[0].boxes) if results else 0
        
        # Store the result
        predictions.append({"image_name": image_name, "predicted_count": predicted_count})
    else:
        print(f"Image not found: {image_name}")

# Save predictions to CSV
with open(output_csv, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["image_name", "predicted_count"])
    writer.writeheader()
    writer.writerows(predictions)

print(f"Predictions saved to {output_csv}")

# Load ground truth counts
ground_truth = {}
with open(ground_truth_csv, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        ground_truth[row['image_name']] = int(row['total_count'])

# Evaluate accuracy
total_images = 0
total_absolute_error = 0
total_ground_truth = 0
total_predictions = 0

for prediction in predictions:
    image_name = prediction["image_name"]
    predicted_count = prediction["predicted_count"]
    gt_count = ground_truth.get(image_name, 0)
    
    total_images += 1
    total_absolute_error += abs(predicted_count - gt_count)
    total_ground_truth += gt_count
    total_predictions += predicted_count

# Calculate metrics
mean_absolute_error = total_absolute_error / total_images if total_images > 0 else 0
accuracy = 1 - (total_absolute_error / total_ground_truth) if total_ground_truth > 0 else 0

print(f"Evaluation Results:")
print(f"Mean Absolute Error (MAE): {mean_absolute_error:.2f}")
print(f"Accuracy: {accuracy:.2%}")

# Save evaluation metrics
evaluation_csv = r"C:\Users\USER\Desktop\ant\evaluation_results_2.csv"
with open(evaluation_csv, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Metric", "Value"])
    writer.writerow(["Mean Absolute Error (MAE)", mean_absolute_error])
    writer.writerow(["Accuracy", f"{accuracy:.2%}"])

print(f"Evaluation metrics saved to {evaluation_csv}")
```

```python
from ultralytics import YOLO
import os
import csv
import cv2  # OpenCV for drawing bounding boxes
import numpy as np

# Paths
model_path = r"C:\Users\USER\Desktop\ant\runs\detect\new\additional_training\weights\epoch100.pt"
penguin_list_path = r"C:\Users\USER\Desktop\ant\ant_image_names.txt"
image_dir = r"C:\Users\USER\Downloads\FSC147_384_V2\images_384_VarV2"
output_csv = r"C:\Users\USER\Desktop\penguin\yolopredict.csv"
ground_truth_csv = r"C:\Users\USER\Desktop\ant\ant_gt_counts.csv"
output_image_dir = r"C:\Users\USER\Desktop\penguin\output_images"  # Directory to save annotated images

# Ensure output directory exists
os.makedirs(output_image_dir, exist_ok=True)

# Load the YOLO model
model = YOLO(model_path)

# Load the list of penguin image names
with open(penguin_list_path, 'r') as file:
    penguin_image_names = [line.strip() for line in file]

# Predict penguin counts and save results
predictions = []

for image_name in penguin_image_names:
    image_path = os.path.join(image_dir, image_name)
    
    if os.path.exists(image_path):
        # Perform prediction
        results = model.predict(source=image_path)
        
        # Count detected penguins and extract bounding boxes
        predicted_count = 0
        if results:
            predicted_count = len(results[0].boxes)
            bounding_boxes = results[0].boxes.xyxy.cpu().numpy()  # Fix: Convert to numpy after moving to CPU
        else:
            bounding_boxes = []
        
        # Store the result
        predictions.append({"image_name": image_name, "predicted_count": predicted_count})
        
        # Load the image
        image = cv2.imread(image_path)
        
        # Draw bounding boxes on the image
        for box in bounding_boxes:
            x1, y1, x2, y2 = map(int, box[:4])  # Convert to integers
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box for prediction
            cv2.putText(image, "Penguin", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        
        # Save the annotated image
        output_image_path = os.path.join(output_image_dir, image_name)
        cv2.imwrite(output_image_path, image)
    else:
        print(f"Image not found: {image_name}")

# Save predictions to CSV
with open(output_csv, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["image_name", "predicted_count"])
    writer.writeheader()
    writer.writerows(predictions)

print(f"Predictions and annotated images saved to {output_csv} and {output_image_dir}")
```
